```python
import graphviz as gr
```

```python
dot = gr.Digraph('round-table', comment='The round table')

dot.node('A', 'King Arthur')
dot.node('B', 'Sir Bedevere the Wise')
dot.node('L', 'Sir the Lancelot the Brave')

dot.edges(['AB','AL'])
dot.edge('B','L',constraint = 'false')
dot
```
